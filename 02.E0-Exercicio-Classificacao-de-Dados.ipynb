### Seleção de algoritmos de classificação

Nessa etapa selecionamos estes algoritmos:
- Naive Bayes
- k-Nearest Neighbors
- Support Vector Machine (Linear/RBF)
- Decision Trees
- Random Forest
- Multilayer Perceptron
- Logistic Regression
- XGBoost

#### Naive Bayes
Baseia-se no teorema de Bayes que é uma equação para calcular a relação de probabilidades condicionais de uma instância pertencer a uma classe.

#### k-Nearest Neighbors

O k-NN classifica pontos de dados com base na maioria das classes dos k pontos de treinamento mais próximos. Ele utiliza a distância entre os pontos de dados para determinar a classe.

#### Support Vector Machine

Procura encontrar o hiperplano que melhor separa as classes no espaço de características. O SVM pode usar diferentes tipos de kernels, como o linear (para separação linear) e o RBF (para separação não linear).

#### Decision Trees

Utilizam uma estrutura de árvore para tomar decisões com base nas características dos dados. Cada nó da árvore representa uma decisão com base em uma característica, e as folhas representam as classes finais.

#### Random Forest

Combina várias árvores de decisão. Ele treina múltiplas árvores em subconjuntos aleatórios dos dados e combina as previsões para obter um resultado mais robusto.

#### Multilayer Perceptron

Rede neural artificial com múltiplas camadas de neurônios. Ele consiste em várias camadas de neurônios interconectados, onde cada camada processa e transforma os dados.

#### Logistic Regression

Estima a probabilidade de um ponto de dados pertencer a uma classe. Ele modela a relação entre as variáveis de entrada e a probabilidade de pertencer a uma classe específica usando uma função logística.

### Treinamento e avaliação

### Treinamento do Naive Bayes

Treinando o algoritmo usando todo o conjunto de treinamento.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

model_nb = GaussianNB()
model_nb.fit(Xtrain, ytrain)

predictions_nb = model_nb.predict(Xtest)
predictions_nb
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier()
model_knn.fit(Xtrain, ytrain)

predictions_knn = model_knn.predict(Xtest)
predictions_knn
```

### Treinamento do SVM

Treinando o algoritmo usando todo o conjunto de treinamento.

```python
from sklearn.svm import SVC

model_svc_linear = SVC(kernel='linear')
model_svc_rbf = SVC(kernel='rbf')

model_svc_linear.fit(Xtrain, ytrain)
model_svc_rbf.fit(Xtrain, ytrain)

predictions_svc_linear = model_svc_linear.predict(Xtest)
predictions_svc_rbf = model_svc_rbf.predict(Xtest)
```

### Treinamento do Multilayer Perceptron

```python
from sklearn.neural_network import MLPClassifier

model_mlp = MLPClassifier(max_iter=1000)

model_mlp.fit(Xtrain, ytrain)

predictions_mlp = model_mlp.predict(Xtest)
predictions_mlp
```

### Treinamento do Logistic Regression

Treinando o algoritmo usando todo o conjunto de treinamento.

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(solver='lbfgs', max_iter=10000)

model_lr.fit(Xtrain, ytrain)

predictions_lr = model_lr.predict(Xtest)
predictions_lr
```

### Avaliação do desempenho usando métricas como acurácia, precisão, recall e F1-score.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model, ytest, predictions):
    accuracy = accuracy_score(ytest, predictions)
    precision = precision_score(ytest, predictions, zero_division=1)
    recall = recall_score(ytest, predictions, zero_division=1)
    f1 = f1_score(ytest, predictions, zero_division=1)
    
    print(f"{model} Metrics:")
    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1: {f1}")
    print()

model_predictions = {
    "Naive Bayes": predictions_nb,
    "K-Nearest Neighbors": predictions_knn,
    "SVC Linear": predictions_svc_linear,
    "SVC RBF": predictions_svc_rbf,
    "Decision Trees": predictions_dtree,
    "Random Forest": predictions_rf,
    "Multilayer Perceptron": predictions_mlp,
    "Logistic Regression": predictions_lr
}

for name, predictions in model_predictions.items():
    evaluate_model(name, ytest, predictions)
```

### Validação cruzada

Aplicando a validação cruzada para cada um dos modelos utilizados no trabalho.

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, shuffle=True)

def cross_validation(model, Xtrain, ytrain):
    scores = cross_val_score(model, Xtrain, ytrain, cv=kfold)
    print(f"{model.__class__.__name__}: {scores.mean()}")
    print()

models = [model_nb, model_knn, model_svc_linear, model_svc_rbf, model_dtree, model_rf, model_mlp, model_lr]

for model in models:
    cross_validation(model, Xtrain, ytrain)
    predictions = model.predict(Xtest)
    evaluate_model(model.__class__.__name__, ytest, predictions) 
```

### Ajuste de hiperparâmetros

```python
from sklearn.model_selection import GridSearchCV

models = [
    ('Naive Bayes', model_nb, {'var_smoothing': [1e11, 1e-10, 1e-9]}),
    ('K-Nearest Neighbors', model_knn, {'n_neighbors' : [1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]}),
    ('SVM', model_svc_linear, {'kernel': ['linear', 'rbf', 'poly', 'sigmoid']}),
    ('Decision Tree', model_dtree, {'max_depth': [5, 10, 15]}),
    ('Random Forest', model_rf, {'n_estimators': [100, 200, 300]}),
    ('Multilayer Perceptron', model_mlp, {'activation': ['tanh', 'relu'], 'solver': ['sgd', 'adam']}),
    ('Logistic Regression', model_lr, {'C': [100, 10, 1.0, 0.1, 0.01]}),
]

for name, model, param in models:
    grid_search = GridSearchCV(model, param, cv=5)
    grid_search.fit(Xtrain, ytrain)
    
    display(grid_search.best_params_)
    display(grid_search.best_score_)

    predictions = model.predict(Xtest)
    evaluate_model(model.__class__.__name__, ytest, predictions)
```
